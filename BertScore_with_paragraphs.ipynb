```python
def evaluate_all_summaries(paragraphs, summaries, output_csv):
    results = []

    for i, paragraph in enumerate(paragraphs):
        paragraph_results = []
        for j, summary in enumerate(summaries[i]):
            # Calculate BERTScore for each summary
            _, _, F = score([summary], [paragraph], lang="en", verbose=False)
            paragraph_results.append(F.item())
        results.append(paragraph_results)

    with open(output_csv, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Paragraph', 'Summary', 'BERTScore'])

        for i, paragraph_result in enumerate(results):
            for j, f_score in enumerate(paragraph_result):
                writer.writerow([i + 1, j + 1, f_score])
```
